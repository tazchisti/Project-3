```python
import pandas as pd
```

```python
path = "Defense/defense.csv"
data = pd.read_csv(path,encoding="ISO-8859-1")
data.head()
```

```python
new = data["Drafted"]= data["Drafted"].str.split("/", n = 3, expand = True)
data["Round"] = new[1]
data["Pick"] = new[2]
data.head()
```

```python
new_1 = data["Round"]= data["Round"].str.split("th", n = 1, expand = True)
data["Round"] = new_1[0]
data.head(50)
```

```python
new_2 = data["Round"]= data["Round"].str.split("nd", n = 1, expand = True)
data["Round"] = new_2[0]
data.head(50)
```

```python
new_3 = data["Round"]= data["Round"].str.split("st", n = 1, expand = True)
data["Round"] = new_3[0]
data.head(50)
```

```python
new_4 = data["Round"]= data["Round"].str.split("rd", n = 1, expand = True)
data["Round"] = new_4[0]
data.head(50)
```

```python
data.drop(columns =["Pick"], inplace = True) 
data.head()
```

```python
data.to_csv("Defense/clean_defense.csv", index=False, header=True)
```
